```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from params import subjects, stages_labels
```

```python
destination_folder = f'../resp_stats/'
```

```python
concat_rsp_features = []
rows_global = []
rows_by_stage = []
for subject in subjects:
    df_subject = pd.read_excel(f'../resp_features/{subject}_resp_features_tagged.xlsx', index_col = 0)
    n_tot = df_subject.shape[0]
    n_by_stage = df_subject['sleep_stage'].value_counts()
    tot_spindled = df_subject['Spindle_Tag'].sum()
    prop_spindled = tot_spindled / n_tot
    tot_slowwaved = df_subject['SlowWave_Tag'].sum()
    prop_slowwaved = tot_slowwaved / n_tot
    rows_global.append([subject, n_tot, tot_spindled, prop_spindled, tot_slowwaved, prop_slowwaved])

    for stage in stages_labels:
        df_subject_stage = df_subject[df_subject['sleep_stage'] == stage]
        n_in_stage = df_subject_stage.shape[0]
        prop_in_stage = n_in_stage / n_tot
        n_spindled = df_subject_stage['Spindle_Tag'].sum()
        prop_spindled = n_spindled / n_in_stage
        n_slowwaved = df_subject_stage['SlowWave_Tag'].sum()
        prop_slowwaved = n_slowwaved / n_in_stage
        rows_by_stage.append([subject, stage, n_in_stage , prop_in_stage, n_spindled, prop_spindled, n_slowwaved, prop_slowwaved])
        
    
    df_subject.insert(0, 'subject', subject)
    concat_rsp_features.append(df_subject)
        
global_stats = pd.DataFrame(rows_global, columns = ['subject', 'n cycles', 'n spindled', 'proportion spindled', 'n slowwaved', 'proportion slowwaved'])
staged_stats = pd.DataFrame(rows_by_stage, columns = ['subject', 'stage', 'n in stage', 'proportion in stage', 'n spindled in stage', 'proportion spindled in stage', 'n slowwaved in stage', 'proportion slowwaved in stage'])
pooled_features = pd.concat(concat_rsp_features)

```

```python
global_stats.to_excel(destination_folder + 'global_rsp_events_stats.xlsx')
staged_stats.to_excel(destination_folder + 'rsp_stage_events_stats.xlsx')
```

```python
fig, ax = plt.subplots()
global_stats[['n cycles','n spindled','n slowwaved']].sum().plot.bar(ax=ax)
ax.set_title('Pooled resp cycles from all subjects')
plt.savefig(destination_folder + 'pooling_cycles_N', bbox_inches = 'tight')
plt.show()
```

```python
fig, axs = plt.subplots(ncols = 2, figsize =(15,5))
fig.suptitle('Proportion of events in respiratory cycles by stage')
ax = axs[0]
sns.boxplot(data = staged_stats, x = 'stage', y = 'proportion spindled in stage', ax=ax)
ax = axs[1]
sns.boxplot(data = staged_stats, x = 'stage', y = 'proportion slowwaved in stage', ax=ax)
plt.savefig(destination_folder + 'boxplot_proportion_event_by_stage', bbox_inches= 'tight')
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 3, figsize = (15,8), constrained_layout = True)
fig.suptitle('Effect of event presence in resp cycles on their features')
for r, predictor in enumerate(['Spindle_Tag', 'SlowWave_Tag']):
    for c, outcome in enumerate(['cycle_freq', 'inspi_volume', 'expi_volume']):
        ax = axs[r,c]
        sns.violinplot(data=pooled_features,  x= predictor, y = outcome, ax=ax)
plt.savefig(destination_folder + 'violin_eventing_effect', bbox_inches= 'tight')
```

```python
outcomes = np.array(['cycle_freq','cycle_ratio','inspi_volume','expi_volume']).reshape(2,2)
fig, axs = plt.subplots(nrows = 2, ncols = 2, constrained_layout = True, figsize = (15,8))
fig.suptitle('Effect of sleep staging on respiratory features')
for r in range(2):
    for c in range(2):
        ax = axs[r,c]
        outcome = outcomes[r,c]
        sns.violinplot(data = pooled_features, x = 'sleep_stage', y = outcome ,ax=ax)
plt.savefig(destination_folder + 'plot_sleep_stage_effect', bbox_inches= 'tight')
```

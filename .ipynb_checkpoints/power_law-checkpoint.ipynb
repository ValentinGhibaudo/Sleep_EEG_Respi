```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import ghibtools as gh
```

```python
srate =256
```

```python
da = xr.load_dataarray('../dataarray/da_N2N3_P2.nc')
```

```python
sig = da.loc['C4-T4',0:300]
```

```python
sig.plot()
```

```python
tf = gh.tf(sig, srate= 256, f_start = 1 , f_stop = 101, n_step = 200, cycle_start = 3 , cycle_stop = 30)
```

```python
modes = ['decibel','prctchange','ztransform','divide']
concat = []
for mode in modes:
    tf_scaled = gh.tf_power_law(tf, 30, 30.2, method = mode)
    concat.append(tf_scaled)
tf_scaled_methods = xr.concat(concat, dim = 'method').assign_coords({'method':modes})
```

```python
for method in modes:
    tf_scaled_methods.loc[method,:,30:31].plot.pcolormesh(x='time',y='freqs', robust = True, yscale='linear')
    plt.show()
```

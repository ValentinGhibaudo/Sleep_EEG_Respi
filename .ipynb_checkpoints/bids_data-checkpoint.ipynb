```python
import mne
import mne_bids
from mne_bids import BIDSPath, write_raw_bids, read_raw_bids, print_dir_tree, make_report, inspect_dataset
from mne_bids.stats import count_events
import glob
```

```python
patients = ['P1','P2','P3','P4','P5','P6','P7','P8','P9','P10']
bids_root = '../bids_data/'
```

```python
for participant in patients:
    subject = participant
    data_file = glob.glob(f'../data/{subject}/*.edf')[0]
    bids_path = BIDSPath(subject=subject, task = 'sleep', datatype='eeg', root = bids_root) # create a BIDS object
    raw = mne.io.read_raw_edf(data_file, eog = ['EOGDt','EOGG'], misc = ['ECG','Menton','JBdt','JBg','THERM','SAT','THOD','ABD','PTT','FC','Pouls','POS','Micro','PCO2','DEBIT','CPAP','Pression'], exclude = ['Pouls','POS','Micro','CPAC','Pression'])
    write_raw_bids(raw, bids_path, overwrite=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import yasa
import seaborn as sns
import ghibtools as gh
import mne
import glob
from params import *
```

```python
def get_stage_duration(sleep_stats, subject, stage):
    return sleep_stats.set_index('subject').loc[subject, stage]
```

```python
sleep_stats = pd.read_excel(f'../subject_characteristics/global_sleep_stats.xlsx', index_col = 0)
rows = []
evs = ['spindles','slowwaves']
for ev in evs:
    for sub in subjects:
        ev_masto = pd.read_excel(f'../event_detection/{sub}_{ev}_reref.xlsx', index_col = 0)
        ev_bipol = pd.read_excel(f'../event_detection/{sub}_{ev}_bipol.xlsx', index_col = 0)
        modes = {'bipol':ev_bipol, 'masto':ev_masto}
        for mode in modes.keys():
            df = modes[mode]
            n_ev_total = df.shape[0]
            for stage in ['N2','N3']:
                df_stage = df[df['Stage_Letter'] == stage]
                n_ev_stage = df_stage.shape[0]
                stage_duration = get_stage_duration(sleep_stats, sub, stage)
                density_by_stage = n_ev_stage / stage_duration
                for chan in df_stage['Channel'].unique():
                    df_stage_chan = df_stage[df_stage['Channel'] == chan]
                    n_ev_stage_chan = df_stage_chan.shape[0]
                    density_by_stage_by_chan = n_ev_stage_chan / stage_duration

                    row = [sub, ev, mode, stage, chan, n_ev_total, stage_duration, n_ev_stage, density_by_stage, n_ev_stage_chan, density_by_stage_by_chan ]
                    rows.append(row)
df_spindles_comparison = pd.DataFrame(rows, columns = ['subject','event','mode','stage','chan','n_events_total','stage_duration','n_events_stage','density_by_stage','n_events_stage_chan','density_by_stage_by_chan'])
```

```python
df_spindles_comparison
```

```python
fig, axs = plt.subplots(ncols = 2, figsize = (15,5), sharey = True)
fig.subplots_adjust(wspace = 0)
for col, mode in enumerate(['bipol','masto']): 
    ax = axs[col]
    N_mean = df_spindles_comparison.groupby(['mode']).mean().loc[mode,'n_spindles_total']
    df_plot = df_spindles_comparison[df_spindles_comparison['mode'] == mode]
    sns.pointplot(data = df_plot , x = 'chan', y= 'density_by_stage_by_chan', hue = 'stage', ax=ax)
    ax.set_title(f'{mode} - N total moyen par sujet : {int(N_mean)}')
    ax.set_ylim(0,4)
    if col == 0:
        ax.set_ylabel('DensitÃ© de fuseaux par minute / stade / chan')
    else:
        ax.set_ylabel(None)
plt.savefig('../autres/ref_mode_spindles')
```
